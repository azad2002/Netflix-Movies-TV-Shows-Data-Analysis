## Phase 3: Data Cleaning and Transformation

This phase focused on improving data quality by handling missing values and addressing inconsistencies in textual variables. Missing values in textual columns were treated using descriptive placeholders such as “Unknown” to avoid losing a large number of records and to preserve data completeness.

Additionally, textual noise and inconsistencies were observed in certain fields, particularly in the title column, due to encoding-related issues that resulted in the appearance of incorrect characters. These issues were addressed through text cleaning and formatting improvements to enhance data clarity and support more accurate analysis and aggregation.

Overall, these preprocessing steps contributed to improving data reliability and reducing textual noise, resulting in a cleaner dataset that is more suitable for subsequent analytical phases.

```python
df.isnull().sum()
```

```python
print(df.duplicated().sum())
```

```python
df.head(10)
```

```python
df.columns
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Listed')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unrated')
```

```python
df.isnull().sum()
```

```python
df['date_added'] = df['date_added'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['date_added']
```

```python
import re

def clean_title(text):
    if pd.isna(text):
        return text
    text = str(text)
    text = text.encode('utf-8', 'ignore').decode('utf-8')
    text = re.sub(r'[\n\r\t]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.strip()
    return text

df['title'] = df['title'].apply(clean_title)
```

```python

df['title'] = df['title'].str.strip()
```

```python
df['title'] = df['title'].str.title()
```

```python

df['title'] = (
   df['title']
    .str.replace(r'\s+', ' ', regex=True)
    .str.replace(r'[^\w\s]', '', regex=True)
)
```

```python
df = df.drop_duplicates(subset='title')
```

## Phase 6: Data Visualization with Matplotlib

This chart illustrates the overall distribution of content types available on the Netflix platform, showing that the number of movies exceeds the number of TV shows. A bar chart was selected as it is the most appropriate visualization for directly and clearly comparing record counts across different categorical groups.

```python
plt.figure(figsize=(6,4))
content_count.plot(kind='bar')
plt.title('Distribution of Content Types on Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Content Type',fontsize=12)
plt.ylabel('Number of Titles',fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('content_type_distribution.png')
plt.show()
```

This chart presents the distribution of content release years and reveals a clear concentration in more recent years, particularly after 2010. A histogram was used because it effectively displays the shape of the distribution and the spread of numerical values over time.

```python
plt.figure(figsize=(7,4))
plt.hist(df['release_year'], bins=20)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(axis='y', alpha=0.3)
plt.legend(fontsize=10)
plt.tight_layout()

plt.savefig('release_year_distribution.png')
plt.show()
```

This chart depicts the temporal trend of content additions to the Netflix platform, highlighting a noticeable increase in the number of titles added in recent years. A line chart was chosen as it is best suited for representing changes over time and clearly emphasizing overall trends.

```python
content_over_time = df.groupby('year_added')['show_id'].count()

plt.figure(figsize=(8,4))
plt.plot(content_over_time)
plt.title('Number of Titles Added to Netflix Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('content_over_time.png')
plt.show()
```

## Phase 7: Data Visualization with Seaborn

This chart presents the statistical distribution of release years along with a density curve, revealing a clear skew toward more recent years. The KDE curve provides a smoother view of the distribution compared to raw numbers or a histogram alone and highlights the strong concentration of content within a specific time period.

```python
plt.figure(figsize=(8,4))
sns.histplot(df['release_year'], kde=True, bins=20)
plt.title('Distribution of Release Years with Density')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

This visualization illustrates the relative distribution of Netflix content by primary genre. A clear concentration of content appears in a small number of genres, while the rest are represented by progressively smaller counts. The visual representation highlights the magnitude gap between different genres, a pattern that is less evident when viewing numerical values alone. This distribution reflects Netflix’s tendency to prioritize specific genres that meet global audience demand, while maintaining limited diversity in less common genres.

```python
genre_counts = (
    df.groupby('primary_genre')['show_id'].count().sort_values(ascending=False).reset_index().rename(columns={'show_id': 'count'})
)

sns.set_style("whitegrid")

plt.figure(figsize=(12,6))
sns.barplot(
    data=genre_counts,
    x='primary_genre',
    y='count',
    palette='viridis'
)

plt.title("Distribution of Netflix Titles by Primary Genre", fontsize=14)
plt.xlabel("Primary Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('dis of')
plt.show()
```

This chart reveals differences in temporal production patterns between movies and TV shows. A clear increase in content production can be observed in recent years, with noticeable differences in distribution density between the two content types—patterns that are not apparent when each category is analyzed separately.

```python
sns.displot(
    data=df,
    x='release_year',
    col='type',
    bins=20,
    kde=True,
    palette='coolwarm',
    height=4,
    aspect=1.2
)

plt.suptitle("Content Release Trend Over Years by Type", y=1.05)
plt.show()
```

This chart shows that Netflix content is heavily concentrated in a limited number of primary genres. The most dominant genres account for a large proportion of the total content, while the remaining genres gradually decline in volume. This pattern reflects a strategic focus on genres with the highest audience demand.

```python
top_genres = genre_counts.head(10)

plt.figure(figsize=(12,6))
sns.barplot(
    data=top_genres,
    x='primary_genre',
    y='count',
    palette='viridis'
)

plt.title("Top 10 Primary Genres on Netflix")
plt.xlabel("Primary Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Phase 8: Synthesis and Interpretation of Findings

The complete analysis of the Netflix dataset provides a comprehensive understanding of the platform’s content structure, quality, and strategic direction. By integrating data exploration, cleaning, feature engineering, aggregation, and visualization, clear and consistent patterns emerged that explain how Netflix organizes and expands its content library.

From Phase 1, the dataset was found to contain 8,807 titles across 12 variables, dominated by categorical attributes such as content type, country, rating, and genre, with release year as the primary numerical variable. Several data quality issues were identified, including missing values in columns such as director, cast, country, date_added, rating, and duration, which could affect content and geographic analysis if not addressed.

In Phase 2, numerical analysis using NumPy highlighted a strong shift toward modern content. The dataset contained 6,216 recent titles and only 2,591 older titles, confirming that Netflix focuses heavily on newer productions. This reflects a strategy aimed at keeping the catalog fresh and aligned with current viewer preferences.

Phase 3 improved data reliability by handling missing values and correcting corrupted text, particularly in titles and cast names affected by encoding issues. Replacing missing categorical values with “Unknown” preserved the completeness of the dataset, while cleaning garbled characters ensured that textual attributes were readable and suitable for analysis.

Phase 4 added analytical depth through feature engineering. New variables such as content age, year added, recent vs. old classification, and primary genre enabled more meaningful comparisons across time and content categories. These engineered features made it possible to study Netflix’s growth patterns and content focus more effectively.

In Phase 5, filtering, grouping, and pivot tables were used to generate meaningful summaries. The results showed that movies outnumber TV shows, and that content distribution varies significantly across genres. Pivot tables also revealed a strong increase in content additions over recent years, highlighting Netflix’s aggressive expansion strategy.

Phase 6 visualized these findings using bar charts, histograms, and line plots. These visuals clearly illustrated the dominance of movies, the concentration of release years after 2010, and the rapid growth in content additions. Visual representation made these trends far more intuitive than numerical tables alone.

Phase 7 provided deeper statistical insight using Seaborn. Genre distribution plots revealed that Netflix’s content is concentrated in a small number of dominant genres, with Drama being the most common genre, followed by Comedy. This indicates a strong strategic focus on high-demand genres that attract broad global audiences. KDE and distribution plots further confirmed a strong skew toward recent years, while comparative visualizations between movies and TV shows revealed different production and release patterns that were not apparent from summary statistics alone.

Overall, the combined findings demonstrate that Netflix follows a focused, data-driven content strategy. The platform prioritizes recent productions and highly popular genres such as Drama and Comedy, while still maintaining some diversity through less frequent categories. This balanced approach allows Netflix to maximize viewer engagement, remain competitive, and continuously adapt to global audience demand in a rapidly evolving streaming market.
